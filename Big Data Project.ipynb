### Preprocessing
Remove Unnecessary Columns in MongoDB

```python
sample_data = list(collection.find({}, {"_id": 0, "poster_path": 1, "backdrop_path": 1, "homepage": 1, "tagline": 1}).limit(10))

# Convert to DataFrame
df_before = pd.DataFrame(sample_data)

# Display the DataFrame
print("üîπ Data BEFORE Removing Unnecessary Fields:")
df_before.head(10)
```

```python
# Remove unnecessary columns
update_result = collection.update_many({}, { 
    "$unset": { 
        "poster_path": 1, 
        "backdrop_path": 1, 
        "homepage": 1, 
        "tagline": 1
    } 
})

# Print update summary
print(f"Matched Documents: {update_result.matched_count}")
print(f"Modified Documents: {update_result.modified_count}")
```

Convert Boolean Fields from string to Boolean

```python
# Fetch sample records with boolean fields
sample_boolean_data = list(collection.find({}, {"_id": 0, "adult": 1, "in_production": 1}).limit(10))

# Convert to DataFrame
df_boolean_before = pd.DataFrame(sample_boolean_data)

# Display DataFrame before boolean conversion
print("üîπ Data BEFORE Converting Boolean Strings:")
display(df_boolean_before)
```

```python
# Convert boolean fields in MongoDB
update_result = collection.update_many({}, [
    { "$set": { 
        "adult": { "$cond": { "if": { "$eq": ["$adult", "True"] }, "then": True, "else": False } },
        "in_production": { "$cond": { "if": { "$eq": ["$in_production", "True"] }, "then": True, "else": False } }
    }}
])

# Print update summary
print(f"‚úÖ Matched Documents: {update_result.matched_count}")
print(f"‚úÖ Modified Documents: {update_result.modified_count}")
```

Convert Date Strings (first_air_date, last_air_date) to ISODate

```python
# Fetch sample records with date fields
sample_date_data = list(collection.find({}, {"_id": 0, "first_air_date": 1, "last_air_date": 1}).limit(10))

# Convert to DataFrame
df_date_before = pd.DataFrame(sample_date_data)

# Display DataFrame before date conversion
print("üîπ Data BEFORE Converting Date Strings:")
display(df_date_before)
```

```python
# Convert date fields while handling empty values
update_result = collection.update_many(
    { "$and": [ { "first_air_date": { "$ne": "" } }, { "first_air_date": { "$exists": True } } ] },
    [{ "$set": { "first_air_date": { "$toDate": "$first_air_date" } } }]
)

update_result_2 = collection.update_many(
    { "$and": [ { "last_air_date": { "$ne": "" } }, { "last_air_date": { "$exists": True } } ] },
    [{ "$set": { "last_air_date": { "$toDate": "$last_air_date" } } }]
)

# Print update summary
print(f"‚úÖ Matched and Modified `first_air_date`: {update_result.modified_count}")
print(f"‚úÖ Matched and Modified `last_air_date`: {update_result_2.modified_count}")
```

 Convert genres from a String to an Array

```python
# Fetch sample records with genres field
sample_genres_data = list(collection.find({}, {"_id": 0, "genres": 1}).limit(10))

# Convert to DataFrame
df_genres_before = pd.DataFrame(sample_genres_data)

# Display DataFrame before conversion
print("üîπ Data BEFORE Converting `genres` to Array:")
display(df_genres_before)
```

```python
# Convert `genres` from string to array in MongoDB
update_result = collection.update_many({}, [
    { "$set": { 
        "genres": { 
            "$split": ["$genres", ", "]  # Split by ", " into an array
        }
    }}
])

# Print update summary
print(f"‚úÖ Matched Documents: {update_result.matched_count}")
print(f"‚úÖ Modified Documents: {update_result.modified_count}")
```

Convert episode_run_time to integer and handle nulls

```python
# Fetch sample records with episode_run_time
sample_runtime_data = list(collection.find({}, {"_id": 0, "episode_run_time": 1}).limit(10))

# Convert to DataFrame
df_runtime_before = pd.DataFrame(sample_runtime_data)

# Display DataFrame before conversion
print("üîπ Data BEFORE Converting `episode_run_time`:")
display(df_runtime_before)
```

```python
# Convert `episode_run_time` to Integer while handling nulls
update_result = collection.update_many({}, [
    { "$set": { 
        "episode_run_time": {
            "$convert": {
                "input": "$episode_run_time",
                "to": "int",
                "onError": 0,  # If not convertible, set to 0
                "onNull": 0     # If null, set to 0
            }
        }
    }}
])

# Print update summary
print(f"‚úÖ Matched Documents: {update_result.matched_count}")
print(f"‚úÖ Modified Documents: {update_result.modified_count}")
```

Check and remove duplicate records	

```python
# Find duplicate records by `id`
pipeline = [
    { "$group": { "_id": "$id", "count": { "$sum": 1 }, "docs": { "$push": "$_id" } } },
    { "$match": { "count": { "$gt": 1 } } }
]

# Run aggregation query
duplicates = list(collection.aggregate(pipeline))

# Convert to DataFrame
df_duplicates = pd.DataFrame(duplicates)

# Display duplicate records if any exist
print("üîπ Duplicate Records (Grouped by `id`):")
display(df_duplicates)
```

```python
# Find and remove duplicates while keeping one copy
pipeline = [
    { "$group": { "_id": "$id", "docs": { "$push": "$_id" } } },
    { "$project": { "_id": 1, "keep": { "$arrayElemAt": ["$docs", 0] }, "remove": { "$slice": ["$docs", 1, { "$size": "$docs" }] } } } 
]

# Get the list of duplicate `_id`s to remove
to_remove = []
for doc in collection.aggregate(pipeline):
    to_remove.extend(doc["remove"])

# Remove duplicate records from MongoDB
if to_remove:
    delete_result = collection.delete_many({"_id": {"$in": to_remove}})
    print(f"‚úÖ Removed {delete_result.deleted_count} duplicate records.")
else:
    print("‚úÖ No duplicate records found.")
```

### Analysis

1.	 What are the top 10 highest-rated TV shows?

```python
# Find top 10 highest-rated TV shows
top_rated_shows = list(collection.find(
    {}, 
    {"_id": 0, "name": 1, "vote_average": 1}
).sort("vote_average", -1).limit(10))

# Convert to DataFrame
df_top_rated = pd.DataFrame(top_rated_shows)

# Display the results
print("üîπ Top 10 Highest-Rated TV Shows:")
display(df_top_rated)
```

2.	How many TV shows are produced per country?	

```python
# Query for TV shows by country (excluding "No country available")
tv_shows_by_country_filtered = list(collection.aggregate([
    { "$match": { "origin_country": { "$nin": [["No country available"]] } } },  # Exclude missing countries
    { "$group": { "_id": "$origin_country", "count": { "$sum": 1 } } },
    { "$sort": { "count": -1 } },
    { "$limit": 10 }
]))

# Convert to DataFrame
df_shows_by_country_filtered = pd.DataFrame(tv_shows_by_country_filtered)

# Rename columns for readability
df_shows_by_country_filtered.rename(columns={"_id": "Country", "count": "Number of TV Shows"}, inplace=True)

# Display the top 10 countries excluding missing values
print("üîπ Top 10 TV Show-Producing Countries (Excluding Missing Values):")
display(df_shows_by_country_filtered)
```

3.	What are the most common genres?	

```python
# Query for most common genres (excluding "No genre available")
popular_genres_filtered = list(collection.aggregate([
    { "$unwind": "$genres" },  # Unwind genres array
    { "$match": { "genres": { "$ne": "No genre available" } } },  # Exclude missing genres
    { "$group": { "_id": "$genres", "count": { "$sum": 1 } } },
    { "$sort": { "count": -1 } },
    { "$limit": 10 }
]))

# Convert to DataFrame
df_popular_genres_filtered = pd.DataFrame(popular_genres_filtered)

# Rename columns for readability
df_popular_genres_filtered.rename(columns={"_id": "Genre", "count": "Number of Shows"}, inplace=True)

# Display the top 10 genres excluding missing values
print("üîπ Top 10 Most Common TV Show Genres (Excluding Missing Values):")
display(df_popular_genres_filtered)
```
